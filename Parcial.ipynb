```python
import random, string
import re
dictNumeros = {}
num = []

valueEntered =""
while valueEntered=="":
    valueEntered=input()
    if valueEntered.upper()=='EXIT':
      print("El direccioro de codigos es: ",dictNumeros)
      print("El listado de codigos es: ",num)
    elif valueEntered =="":
        x = ''.join(random.choices(string.ascii_letters, k=1))
        x1 = ''.join(random.choices(string.digits, k=1))
        x2 = ''.join(random.choices(string.ascii_letters, k=1))
        x3 = ''.join(random.choices(string.digits, k=1))
        x4 = ''.join(random.choices(string.ascii_letters, k=1))
        x5 = ''.join(random.choices(string.digits, k=1))
        x6 = ''.join(random.choices(string.digits, k=1))
        x=x+x1+x2+x3+x4+x5+x6
        x=x.upper()
        print(x)
        strNumeros=re.sub(r"[A-z]", "", str(x))
        listaResultado= [i for i in strNumeros]
        
        
        Sumatotal = sum([int(i) for i in listaResultado]) 
        if Sumatotal>=30:
            dictNumeros[x] = Sumatotal
        else:
            num.append(x)
            
        print("estos son los numeros del codigo "+strNumeros)    
        print("Esta es la suma dig numericos "+ str(Sumatotal))
        
    else:
        valueEntered=""
```
